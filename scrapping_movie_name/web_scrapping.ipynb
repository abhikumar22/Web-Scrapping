```python
import requests
import bs4
import pandas as pd
rq = requests.get('https://www.imdb.com/chart/top')
bs = bs4.BeautifulSoup(rq.text , 'lxml')
```

```python
bs.select('.titleColumn' )
```

```python
lst = []
for i in bs.select('.titleColumn'):
    lst += [i.text]
print(lst) 
```

```python
lst1=[]
lst1=[s.replace('\n', '') for s in lst] # remove all the \n
print(lst1)
```

```python
Movies = pd.DataFrame(lst1 , columns = ['Movies'])
Movies
```

```python
Movies.to_csv('Top250_movie_by_ratings.csv' , index = False)
"""Successfully saved the scraped data as a csv file"""
```
