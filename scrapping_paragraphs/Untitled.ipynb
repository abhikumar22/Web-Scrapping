```python
import requests
import bs4
import pandas as pd

d = requests.get("https://en.wikipedia.org/wiki/Blog")
da = bs4.BeautifulSoup(d.text, 'lxml')
```

```python
# da         #this will show full code of the webpage

da.select('p')
```

```python
lst = [i.text for i in da.select('p')]    # select the paragraph area of the webpage
lst
```

```python
lst1 = [i.replace('\n','') for i in lst]
print(lst1)
```

```python
lst1 = [x for x  in lst1 if x]
lst1
```

```python
l = pd.DataFrame(lst1 , columns = ['paragraph'])
l
```

```python
l.to_csv('web_scrape.csv' , index = False)
"""Paragraph saved to csv"""
```
